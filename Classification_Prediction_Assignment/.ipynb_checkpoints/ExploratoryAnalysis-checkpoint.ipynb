```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#read the file
df = pd.read_csv('drop-out.csv')

df.head()
```

```python
#Convert the enroll_date_time to a human interprable format
df['enroll_date_time'] = df.apply(lambda x: dt.datetime.fromtimestamp(int(x[5])), axis = 1)

#enroll_date_time contains many pieces of information.  These are split into year, month, day, hour
#minute and second are extremely unlikely to have an effect on completion, so they are disregarded
df['year'] = df.apply(lambda x: int(str(x[5])[:4]), axis = 1)
df['month'] = df.apply(lambda x: int(str(x[5])[5:7]), axis = 1)
df['day'] = df.apply(lambda x: int(str(x[5])[8:10]), axis = 1)
df['hour'] = df.apply(lambda x: int(str(x[5])[11:13]), axis = 1)
df.drop('enroll_date_time', axis = 1, inplace = True)
```

```python
df.describe(include = 'all')
#No missing data
#years column - Most students are in their first year
#enterance_test_score - distribution is skewed right
#courses_taken - students are enrolling in up to 22 classes in one semester
#complete - about 30% of enrollment instances are dropped
#course_id - seems like there might be two types of courses - ones that begin with 65 and ones that begin with 80
#international & online - most students are neither international nor online
#year & month suggest a relatively even split between enrollment instances before the new year and those after the new year
```

```python
df = pd.get_dummies(df)
df.drop(['complete_no', 'international_no', 'online_no'], axis = 1, inplace = True)
#Convert categorical data to numeric
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(df.corr(), xticklabels = df.columns, yticklabels = df.columns, annot = True)
#Online and international have extremely high correlation
#Student ID has high correlation with other variables, but Student ID should be excluded from analysis anyway
    #Extremely unlikely to have an effect on course completion
#No variable on its own has strong correlation with completion
```
